## **Multi-Feature Logistic Regression Example**

| Age | Salary | Bought Product (1/0) |
| --- | ------ | -------------------- |
| 22  | 25k    | 0                    |
| 25  | 35k    | 0                    |
| 28  | 40k    | 1                    |
| 30  | 45k    | 1                    |
| 35  | 65k    | 1                    |


Why multi-feature?

Because real world problems depend on multiple factors:

- Age

- Salary

- Education

- Gender

- Location

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# Creating 'X' Features
X = np.array([
    [22, 25000],
    [25, 35000],
    [28, 40000],
    [30, 45000],
    [35, 65000]
])

# Creating 'y' Labels/Output
y = np.array([0, 0, 1, 1, 1])

# Fit and Train the model
model = LogisticRegression()
model.fit(X, y)

# Getting Coefficients 
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
# Getting prediction and probability by model
print("Class : ",model.predict([[27, 38000]]))       # 0 or 1
print("Probability : ", model.predict_proba([[27, 38000]])) # probability
```

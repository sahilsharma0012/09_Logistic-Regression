## **Decision Boundary Graph**

Meaning

Decision boundary is the cutting line where: 

p=0.5 (OR) z=0

It separates class 0 vs class 1.

**Example dataset**

| Hours studied | Pass (1) / Fail (0) |
| ------------- | ------------------- |
| 1             | 0                   |
| 2             | 0                   |
| 3             | 0                   |
| 4             | 1                   |
| 5             | 1                   |


```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[1],[2],[3],[4],[5]])
y = np.array([0,0,0,1,1])

model = LogisticRegression()
model.fit(X, y)
```

```python
print(model.coef_)
print(model.coef_[0])
model.coef_[0][0]
```

```python
model.intercept_
```

```python
# Get coefficients
w = model.coef_[0][0]
b = model.intercept_[0]
```

```python
# Decision boundary: w*x + b = 0
boundary = -b / w
print(boundary)
```

```python
# Plot
plt.scatter(X, y, color='blue', label="Data")
plt.axvline(boundary, color='red', label=f"Decision Boundary = {boundary:.2f}")
plt.legend()
plt.show()
```

**Interpretation**

- If x > boundary → class = 1

- If x < boundary → class = 0


# Sigmoid Graph

### **What is sigmoid?**

It takes any number (−∞ to +∞) and "squeezes" it between **0 and 1**.

[
\sigma(z)=\frac{1}{1+e^{-z}}
]

### **Why?**

Because we need probabilities — they must be between 0 and 1.

### **Graph (Intuition)**

* When z is large positive → output near **1**
* When z is large negative → output near **0**
* At z = 0 → output **0.5**

---


```python
import numpy as np
import matplotlib.pyplot as plt

z = np.linspace(-10, 10, 200)
sigmoid = 1 / (1 + np.exp(-z))

plt.plot(z, sigmoid)
plt.xlabel("z (linear value)")
plt.ylabel("sigmoid(z)")
plt.title("Sigmoid Function")
plt.grid(True)
plt.show()
```

### **Meaning in real world**

Example:
If a logistic model produces **z = 2**, probability = 0.88 → Positive class

If z = −3, probability = 0.04 → Negative class
